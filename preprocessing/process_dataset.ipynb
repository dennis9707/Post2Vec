```python
import pandas
# df = pandas.read_csv('../data/csv/Question54.csv',lineterminator='\n')
df = pandas.read_csv('../data/questions/Questions54.csv')
```

```python
df_drop = df.drop("Unnamed: 0", axis=1)
df_drop.head()
```

# Drop NA valus in Title, Body and Tag

```python
# number of rows before
print(f"the initial size of the dataset {df_drop.shape[0]}")
df = df_drop[df_drop['Title'].notna()]
print(f"the size of the dataset after filtering title {df.shape[0]}")

df = df[df['Body'].notna()]
print(f"the size of the dataset after filtering body {df.shape[0]}")

df = df[df['Tags'].notna()]
print(f"the size of the dataset after filtering tag {df.shape[0]}")
# df_drop.head()
```

466800 before, no Nan in title, body and tags
346483 after removing nan in code
346483/466801 = 0.74224990949


```python
# fill NA
df[['Code']] = df[['Code']].fillna("")
```

```python
df.head()
```

# filter spaces in title, body and code

```python
for i, row in df.iterrows():
    title = row['Title']
    # if not isinstance(title, str):
    #     print(df.loc[[i]])
    title = " ".join(title.split())

    text = row['Body']
    text = " ".join(text.split())

    code = row['Code']
    if not code == "":
            code = " ".join(code.split())
    df.at[i,"Title"] = title
    df.at[i,"Body"] = text
    df.at[i,"Code"] = code
```

# use a list to store tags

```python
for i, row in df.iterrows():
    if i == 0:
        continue
    tags = row['Tags']
    tags = tags.replace('<', ' ').replace('>', ' ').strip().split()
    df.at[i,"Tags"] = tags
```
