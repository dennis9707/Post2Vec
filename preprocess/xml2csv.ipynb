```python
import pandas as pd
train = pd.read_csv("../data/tagdc_csv/Posts.csv",encoding = "ISO-8859-1") 
```

```python
train.tail()
```

```python
import ast
tag_cnt_dict = {}
tag_cnt_header = ["tag", "count"]
for index, row in train.iterrows():
    tags = ast.literal_eval(row['Tags'])
    for t in tags:
        if t in tag_cnt_dict:
            tag_cnt_dict[t] += 1
        else:
            tag_cnt_dict[t] = 1
    if index % 10000 == 0:
        print(index)
```

```python
len(tag_cnt_dict)
```

```python
sorted = train.sort_values(by='Id')
```

```python
import json

with open('tag_cnt_dict.txt', 'w') as file:
     file.write(json.dumps(tag_cnt_dict)) 
```

```python
newDict = dict()
# Iterate over all the items in dictionary and filter items which has even keys
for (key, value) in tag_cnt_dict.items():
   # Check if key is even then add pair to new dictionary
   if value >= 50:
       newDict[key] = value
```
